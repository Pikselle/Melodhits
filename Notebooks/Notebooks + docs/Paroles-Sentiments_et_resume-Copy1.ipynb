```python
import pandas as pd
from langdetect import detect
from textblob import TextBlob

# Fonction pour détecter l'émotion dans les paroles
def detect_emotion(lyrics):
    # Vérifier si les paroles sont en anglais
    try:
        if detect(lyrics) != 'en':
            return 'Non anglais'
    except:
        return 'Non anglais'
    
    # Analyser le sentiment des paroles
    blob = TextBlob(lyrics)
    sentiment = blob.sentiment.polarity
    
    # Déterminer l'émotion en fonction du sentiment
    if sentiment < -0.5:
        return 'Insulte'
    elif sentiment > 0.5:
        return 'Amour'
    elif -0.5 <= sentiment < 0:
        return 'Tristesse'
    elif 0 <= sentiment < 0.5:
        return 'Joie'
    else:
        return 'Autre'


# Appliquer la fonction detect_emotion pour ajouter la colonne "Emotion"
df['Emotion'] = df['Paroles'].apply(detect_emotion)

# Enregistrer le dataset mis à jour
df.head(50)

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report


# Diviser les données en ensembles d'entraînement et de test
train_data, test_data, train_labels, test_labels = train_test_split(df['Paroles'], df['Emotion'], test_size=0.2, random_state=42)

# Extraction des caractéristiques avec TF-IDF
vectorizer = TfidfVectorizer()
train_features = vectorizer.fit_transform(train_data)
test_features = vectorizer.transform(test_data)

# Créer le modèle SVM
model = SVC()

# Entraîner le modèle
model.fit(train_features, train_labels)

# Prédictions sur l'ensemble de test
predictions = model.predict(test_features)

# Rapport de classification
report = classification_report(test_labels, predictions)
print(report)
```

```python
import re
import random
from collections import defaultdict

class MarkovLyricsGenerator:
    def __init__(self, corpus='', order=2, length=8):
        self.order = order
        self.length = length
        self.words = re.findall("[a-z']+", corpus.lower())
        self.states = defaultdict(list)

        for i in range(len(self.words) - self.order):
            self.states[tuple(self.words[i:i + self.order])].append(self.words[i + order])
    def gen_sentence(self, length=8, startswith=None):
        terms = None
        if startswith:
            start_seed = [x for x in self.states.keys() if startswith in x]
            if start_seed:
                terms = list(start_seed[0])
        if terms is None:
            start_seed = random.randint(0, len(self.words) - self.order)
            terms = self.words[start_seed:start_seed + self.order]
    
        for _ in range(length):
            if len(terms) >= self.order:
                terms.append(random.choice(self.states[tuple(terms[-self.order:])]))
            else:
                break
    
        return ' '.join(terms)



    def gen_song(self, lines=10, length=8, length_range=None, startswith=None):
        song = []
        if startswith:
            song.append(self.gen_sentence(length=length, startswith=startswith))
            lines -= 1
        for _ in range(lines):
            sent_len = random.randint(*length_range) if length_range else length
            song.append(self.gen_sentence(length=sent_len))
        return '\n'.join(song)
```
